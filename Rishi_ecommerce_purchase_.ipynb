# 4 Highest & Lowest Purchase Prices

```python
highest_price = data['Purchase Price'].max()
lowest_price = data['Purchase Price'].min()

print("Highest Purchase Price:", highest_price)
print("Lowest Purchase Price:", lowest_price)

price_data = pd.DataFrame({'Price Type': ['Highest', 'Lowest'], 'Purchase Price': [highest_price, lowest_price]})
```

```python
plt.figure(figsize=(6,4))
plt.bar(price_data['Price Type'], price_data['Purchase Price'], color=['green', 'red'])
plt.title('Highest & Lowest Purchase Prices (Matplotlib)', fontsize=14)
plt.ylabel('Price')
plt.xlabel('Type')
plt.show()
```

### From the above analyis the highest purchase is of 99.99 and the lowest purchase is of 0

# 7. How many people purchase during the AM & how many people purchase during the PM ?


```python
am_pm_counts = df['AM or PM'].value_counts().reset_index()
am_pm_counts.columns = ['Time of Day', 'Number of Purchases']

print(am_pm_counts.to_markdown(index=False, numalign="left", stralign="left"))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

purchase_counts = df['AM or PM'].value_counts()

plt.figure(figsize=(8, 6))
purchase_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Number of Purchases by Time of Day (AM vs PM)')
plt.xlabel('Time of Day')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=0) 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig
```

### As from the above analysis the people who purchase in PM are 5068 and in AM are 4932 and from this we result we can say that people prefer purchasing most at the PM rather than the AM

# 12. Create a bar chart showing the top 10 companies by number of customers.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Ecommerce Purchases.csv')

company_counts = df['Company'].value_counts()

top_10_companies = company_counts.head(10)

plt.figure(figsize=(12, 6))
top_10_companies.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Top 10 Companies by Number of Customers', fontsize=16)
plt.xlabel('Company', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)

```

# 13. Show a countplot for credit card providers

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Ecommerce Purchases.csv')

provider_counts = df['CC Provider'].value_counts()

plt.figure(figsize=(12, 6))

sns.countplot(data=df,y='CC Provider', order=provider_counts.index,palette='viridis')

plt.title('Count of Purchases by Credit Card Provider')
plt.xlabel('Count of Purchases')
plt.ylabel('Credit Card Provider')
```

### As we can see the Visa 16  digit is the most selling credit card provider with count of purhase around 1750
